```python
import pandas as pd

# abrir o ficheiro de livros (o ";" é para não vir tudo embaralhado)
df = pd.read_csv("livros.csv", sep=";", encoding='latin1')

# dar uma olhadela para ver se os dados vieram certos
print("--- Como os dados estão agora ---")
df.info()
print(df.isnull().sum()) # ver onde falta informação

# limpar os livros que estão com 0 páginas (que é erro)
df_limpo = df[df["paginas"] > 0].copy()

# preencher os anos vazios com o valor do meio
mediana_ano = df_limpo["ano"].median()
df_limpo["ano"] = df_limpo["ano"].fillna(mediana_ano).astype(int)

# criar a coluna de décadas (ex: 1998 vira 1990)
df_limpo["decada"] = (df_limpo["ano"] // 10) * 10

# separar os livros por tamanho (curto, médio ou longo)
def definir_tamanho(p):
    if p < 150: return "Curto"
    elif p <= 350: return "Médio"
    else: return "Longo"

df_limpo["faixa_paginas"] = df_limpo["paginas"].apply(definir_tamanho)

# fazer as contas que o prof pediu
print("\n--- Média de páginas por década ---")
print(df_limpo.groupby("decada")["paginas"].mean())

print("\n--- Os 10 autores que mais aparecem ---")
print(df_limpo["autor"].value_counts().head(10))

# guardar tudo num excel para entregar a aula
df_limpo.to_excel("livros_analisados.xlsx", index=False)
print("\n Já está! O Excel foi criado.")
```

```python
# 1. Aqui vamos tirar os livros com 0 páginas e ver quantos saíram
antes = len(df)
df_limpo = df[df["paginas"] > 0].copy()
depois = len(df_limpo)
print(f"Livros removidos (estavam com 0 páginas): {antes - depois}")

# 2. Arrumar a coluna do ano
# Se o ano estiver vazio, colocamos o valor do meio
mediana_ano = df_limpo["ano"].median()
df_limpo["ano"] = df_limpo["ano"].fillna(mediana_ano).astype(int)

# 3. Criar a coluna de década (ex: 1995 vira 1990)
# A conta "// 10 * 10" arredonda o ano para baixo
df_limpo["decada"] = (df_limpo["ano"] // 10) * 10

# 4. Criar a categoria de tamanho usando lambda
# Curto (<150), Médio (150-350) ou Longo (>350)
df_limpo["faixa_paginas"] = df_limpo["paginas"].apply(
    lambda x: "Curto" if x < 150 else "Médio" if x <= 350 else "Longo"
)

# Ver como a tabela ficou com as colunas novas
print("\nPrimeiras linhas da tabela limpa:")
display(df_limpo.head())
```

```python
# --- NÍVEL 3: HORA DA ANÁLISE ---

# 1. Média de páginas por década
# O groupby junta os livros da mesma época e calcula a média das páginas
media_paginas = df_limpo.groupby("decada")["paginas"].mean().sort_index()
print("Média de páginas por década (os livros estão crescendo?):")
print(media_paginas)

# 2. Os 10 autores que mais aparecem aqui
# O value_counts conta as repetições e o head(10) pega os top 10
top_autores = df_limpo["autor"].value_counts().head(10)
print("\nOs 10 autores com mais livros na nossa lista:")
print(top_autores)

# 3. Oss livros novos (depois de 2010)
# Filtrando os anos novos e contando quantos são Curtos, Médios ou Longos
print("\nTamanho dos livros lançados depois de 2010:")
livros_novos = df_limpo[df_limpo["ano"] > 2010]
print(livros_novos["faixa_paginas"].value_counts())

# 4. Exportar tudo para um arquivo de Excel
# Isso gera o arquivo final para você entregar ao professor
df_limpo.to_excel("livros_analisados.xlsx", index=False)
print("\n✅ TUDO PRONTO! O arquivo 'livros_analisados.xlsx' já pode ser baixado.")
```
