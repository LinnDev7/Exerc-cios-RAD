```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# criando os produtos e os preços [cite: 53-59]
produtos = {
    "Dom Casmurro": ("Literatura", 35.90),
    "O Pequeno Príncipe": ("Infantil", 29.90),
    "Sapiens": ("Ciências", 54.90),
    "Python para Dados": ("Tecnologia", 89.90),
    "Clean Code": ("Tecnologia", 95.00)
}

vendedores = ["Ana Lima", "Carlos Mendes", "Bruno Costa", "Fernanda Rocha"]

# sorteando 50 vendas aleatórias [cite: 60-63]
nomes_prod = np.random.choice(list(produtos.keys()), 50)

dados = {
    "produto": nomes_prod,
    "categoria": [produtos[p][0] for p in nomes_prod],
    "quantidade": np.random.randint(1, 6, 50), # sorteia de 1 a 5 [cite: 66]
    "preco_unit": [produtos[p][1] for p in nomes_prod],
    "vendedor": np.random.choice(vendedores, 50)
}

# criando a tabela no pandas [cite: 68]
df = pd.DataFrame(dados)

# calculando o total de cada linha [cite: 69-70]
df["total_venda"] = df["quantidade"] * df["preco_unit"]

# agrupando por categoria para somar tudo [cite: 73-77]
faturamento = df.groupby("categoria")["total_venda"].sum()

print("Resumo de vendas:")
print(faturamento)

# criando o gráfico de barras simples [cite: 83-93]
faturamento.plot(kind='bar', color='green')
plt.title("Vendas por Categoria")
plt.show()
```
